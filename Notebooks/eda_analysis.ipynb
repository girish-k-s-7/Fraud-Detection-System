# Fraud Detection System :- EDA Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"data/processed/cleaned_fraud.csv")
df.head()
```

**Class Distribution**

```python
class_counts = df['Class'].value_counts()
print(class_counts)

plt.figure(figsize=(5,4))
sns.barplot(x=class_counts.index, y=class_counts.values)
plt.title("Class Distribution (0 = Normal, 1 = Fraud)")
plt.show()
```

The dataset is extremely imbalanced, with fraud cases forming a very small fraction of total transactions.

Accuracy is misleading in such datasets; a model predicting all “Non-Fraud” would still score high.

This makes Recall, Precision, and ROC-AUC more meaningful metrics than accuracy.

**Fraud Analysis**

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Transaction Amount by Class")
plt.show()
```

Fraud transactions typically involve lower median amounts compared to normal transactions.

However, there are occasional high-value fraud outliers.

This shows that fraud detection is not only about large transactions.

**Time Analysis**

```python
df['Hour'] = (df['Time'] // 3600) % 24

plt.figure(figsize=(8,5))
sns.countplot(x='Hour', hue='Class', data=df)
plt.title("Fraud Distribution across Time of Day")
plt.show()
```

Fraud activity is not evenly distributed across the day.

Certain time windows show noticeably higher concentration of fraud.

Time-based features may improve model performance.

**Distribution of Features**

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Amount'], bins=100, kde=True)
plt.title("Transaction Amount Distribution")
plt.xscale("log")
plt.show()
```

Transaction amounts are heavily right-skewed with a long tail.

Most transactions are of low value with a few very large payments.

This justifies the use of log scaling or robust scaling before modeling.

**Correlation Analysis**

```python
corr = df.corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Feature Correlation Matrix")
plt.show()
```

PCA features show very low correlation with each other.

This confirms that dimensionality reduction was already applied before data release.

Feature selection is not required; all PCA components are retained.

**Outliear**

```python
plt.figure(figsize=(6,4))
sns.histplot(fraud['Amount'], bins=50, kde=True)
plt.title("Fraud Amount Distribution")
plt.show()
```

Most fraudulent transactions involve small amounts, not just big losses.
